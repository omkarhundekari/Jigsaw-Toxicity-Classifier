```python
import pandas as pd

df = pd.read_csv("../data/train.csv")
df.shape
```

```python
labels = ["toxic","severe_toxic","obscene","threat","insult","identity_hate"]

df.columns.tolist(), df[labels].sum().sort_values(ascending=False)
```

```python
df["label_count"] = df[labels].sum(axis=1)
df["label_count"].value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

df[labels].sum().sort_values().plot(kind="barh")
plt.title("Jigsaw label counts")
plt.tight_layout()
plt.show()
```
